<h2>Dimensionality Reduction</h2>

```python
# Perform the necessary imports
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import make_pipeline
import matplotlib.pyplot as plt


# Reading data
X = credit_data.drop("Class", axis=1)

# Create scaler: scaler
scaler = StandardScaler()

# Create a PCA instance: pca
pca = PCA()

# Create pipeline: pipeline
pipeline = make_pipeline(scaler,pca)

# Fit the pipeline to 'samples'
pipeline.fit(X)

# Plot the explained variances
features = range(pca.n_components_)
plt.bar(features, pca.explained_variance_)
plt.xlabel('PCA feature')
plt.ylabel('Variance')
plt.xticks(features)
plt.show()
```

<h2>Splitting the dataset</h2>

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix, classification_report
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_new, y, test_size=0.3, random_state=7)
print("Number transactions X_train dataset: ", X_train.shape)
print("Number transactions y_train dataset: ", y_train.shape)
print("Number transactions X_test dataset: ", X_test.shape)
print("Number transactions y_test dataset: ", y_test.shape)
```

<h2>Transform the data</h2>

```python
#scaling using MinMaxScaler or StandardScaler
from sklearn.preprocessing import StandardScaler, MinMaxScaler
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

<h1>Models before class balancing</h1>

<h3>1. Logistic Regression</h3>

```python
# logistic regression object
lr = LogisticRegression()
  
# train the model on train set
lr.fit(X_train_scaled, y_train)
  
lr_predictions = lr.predict(X_test_scaled)
  
# print classification report
print(classification_report(y_test, lr_predictions))
```

<h3>2. Random Forest</h3>

```python
from sklearn.ensemble import  RandomForestClassifier
rfc = RandomForestClassifier()
rfc.fit(X_train_scaled, y_train)

rfc_predictions = rfc.predict(X_test_scaled)
  
# print classification report
print(classification_report(y_test, rfc_predictions))
```

<h3>3. Decision Tree</h3>

```python
from sklearn.tree import DecisionTreeClassifier

dtc = DecisionTreeClassifier(max_depth=4, random_state=4)
dtc.fit(X_train_scaled,y_train)

dtc_predictions = dtc.predict(X_test_scaled)
  
# print classification report
print(classification_report(y_test, dtc_predictions))
```

<h3>4. XGBoost Classifier</h3>

```python
!pip install xgboost
from xgboost.sklearn import XGBClassifier
xgb = XGBClassifier()
xgb.fit(X_train_scaled,y_train)

xgb_predictions = xgb.predict(X_test_scaled)
  
# print classification report
print(classification_report(y_test, xgb_predictions))
```

<h1>Dataset class balancing</h1>



```python
print("Before OverSampling, counts of label '1': {}".format(sum(y_train == 1)))
print("Before OverSampling, counts of label '0': {} \n".format(sum(y_train == 0)))
  
# import SMOTE module from imblearn library
# pip install imblearn (if you don't have imblearn in your system)
!pip install imblearn
from imblearn.over_sampling import SMOTE
sm = SMOTE(random_state = 2)
X_train_res, y_train_res = sm.fit_resample(X_train_scaled, y_train)
  
print('After OverSampling, the shape of train_X: {}'.format(X_train_res.shape))
print('After OverSampling, the shape of train_y: {} \n'.format(y_train_res.shape))
  
print("After OverSampling, counts of label '1': {}".format(sum(y_train_res == 1)))
print("After OverSampling, counts of label '0': {}".format(sum(y_train_res == 0)))
```

<h3>1. Logistic Regression</h3>

```python
lr1 = LogisticRegression()
lr1.fit(X_train_res, y_train_res)
lr1_predictions = lr1.predict(X_test_scaled)
  
# print classification report
print(classification_report(y_test, lr1_predictions))
```

<h3>2. Random Forest</h3>

```python
rfc1 = RandomForestClassifier()
rfc1.fit(X_train_res, y_train_res)

rfc1_predictions = rfc1.predict(X_test_scaled)
  
# print classification report
print(classification_report(y_test, rfc1_predictions))
```

<h3>3. Decision Tree</h3>

```python
dtc1 = DecisionTreeClassifier(max_depth=4, random_state=4)
dtc1.fit(X_train_res, y_train_res)

dtc1_predictions = dtc1.predict(X_test_scaled)
  
# print classification report
print(classification_report(y_test, dtc1_predictions))
```

<h3>4. XGBoost Classifier</h3>

```python
xgb1 = XGBClassifier()
xgb1.fit(X_train_res, y_train_res)

xgb1_predictions = xgb1.predict(X_test_scaled)
  
# print classification report
print(classification_report(y_test, xgb1_predictions))
```
